```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_excel("mahaproject.xlsx")

plt.figure(figsize=(12, 8))

# --- Line Plot ---
plt.subplot(2, 2, 1)
plt.plot(df.index[:500], df['bytes'][:500])
plt.title("Line Plot (Index vs Bytes)")
plt.xlabel("Index")
plt.ylabel("Bytes")

# --- Pie Chart (First 5 values) ---
plt.subplot(2, 2, 2)
pie_values = df['bytes'][:5]
plt.pie(pie_values, labels=df['protocol'][:5].astype(str), autopct="%1.1f%%")
plt.title("Pie Chart (First 5 Rows)")

# --- Bar Chart ---
plt.subplot(2, 2, 3)
plt.bar(df.index[:10], df['bytes'][:10])
plt.title("Bar Chart (10 Samples)")
plt.xlabel("Index")
plt.ylabel("Bytes")

# --- Scatter Plot ---
plt.subplot(2, 2, 4)
plt.scatter(df.index[:500], df['bytes'][:500])
plt.title("Scatter Plot (Index vs Bytes)")
plt.xlabel("Index")
plt.ylabel("Bytes")

# Save dashboard
plt.tight_layout()
plt.savefig("dashboard.png", dpi=300, bbox_inches='tight')
plt.show()
```
